```python
# load all json files in history/

import json
import os

history = []
for file in os.listdir("history"):
    if not file.endswith(".json"):
        continue
    with open(os.path.join("history", file)) as f:
        history.append(json.load(f))

history
```

```python
# sort by timestamp
history.sort(key=lambda x: x["timestamp"])

# convert 'timestamp' to datetime
import datetime

for entry in history:
    entry["date"] = datetime.datetime.fromtimestamp(entry["timestamp"])

history
```

```python
# plot the data

import matplotlib.pyplot as plt

plt.figure(figsize=(20, 10))
plt.plot([entry["date"] for entry in history], [entry["trafficflow"] for entry in history])
```
